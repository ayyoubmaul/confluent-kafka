```python
from confluent_kafka import Producer

import json
from time import sleep
```

```python
def producer_callback(err, msg):
    if err is not None:
        print('Message delivery failed: {}'.format(err))
    else:
        print('Message delivered to {} [{}]'.format(msg.topic(), msg.partition()))
```

```python
producer = Producer(config)
```

### Sent one sample data

```python
s_data = json.dumps(data[0])
print(s_data)
producer.produce("transactions", key="", value=s_data, callback=producer_callback)
# push to broker from buffer zone
producer.flush()
```

## Sent all data

```python
for idx in range(len(data)):

    if idx % 2 == 0 :
        sleep(1)

    s_data = json.dumps(data[idx])
    producer.produce("transactions", key="", value=s_data, callback=producer_callback)
    producer.flush()
```
